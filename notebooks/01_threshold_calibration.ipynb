# Threshold Calibration
Computes precision–recall curves over labelled pairs and emits the auto-merge / review thresholds the matching engine uses in production.

```python
import numpy as np, json
from services.matching_engine.thresholds.calibrator import calibrate
scores = np.concatenate([np.random.beta(8,2,200), np.random.beta(2,8,200)])
labels = np.array([1]*200 + [0]*200)
t = calibrate(scores.tolist(), labels.tolist(), target_precision=0.99)
print(json.dumps(t.__dict__, indent=2))
```
